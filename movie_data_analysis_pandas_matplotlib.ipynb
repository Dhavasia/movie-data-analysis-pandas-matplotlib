```python
movies_per_year = df['year'].value_counts().sort_index()
plt.figure(figsize=(10,6))
movies_per_year.plot(kind='line', marker='o', color='orange')
plt.title('Number of Movies Released per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.tight_layout()
plt.show()
```
